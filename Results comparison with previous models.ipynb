```python
import numpy as np
import matplotlib.pyplot as plt

# Time vector
t = np.linspace(0, 200, 1000)

# =======================
# Your HBV Model (With Awareness)
# =======================
A = 200 * (1 - np.exp(-0.05 * t)) * np.exp(-0.01 * t)  # Aware
S = np.exp(-0.01 * t) * 1000                            # Susceptible
V = 250 * (1 - np.exp(-0.03 * t))                      # Vaccinated
Ac = 50 * np.exp(-0.02 * t) * np.sin(0.05 * t)**2      # Acute infections
C = 40 + 10 * np.exp(-0.005 * t)                       # Chronic infections

# =======================
# Previous HBV Model (No Awareness)
# =======================
A_prev = np.zeros_like(t)                              # No awareness component
S_prev = np.exp(-0.005 * t) * 1000
V_prev = 250 * (1 - np.exp(-0.01 * t))                 # Less effective vaccination
Ac_prev = 60 * np.exp(-0.01 * t) * np.sin(0.04 * t)**2
C_prev = 50 + 15 * np.exp(-0.004 * t)

# =======================
# Plot 1: Awareness Comparison
# =======================
plt.figure(figsize=(10, 5))
plt.plot(t, A, label='Aware (The Model)', color='orange')
plt.plot(t, A_prev, label='Aware (Previous Model)', linestyle='--', color='gray')
plt.title('Effect of Awareness on HBV Transmission')
plt.xlabel('Time (days)')
plt.ylabel('Population')
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# =======================
# Plot 2: Vaccination Impact
# =======================
plt.figure(figsize=(10, 5))
plt.plot(t, S, label='Susceptible (The Model)', color='blue')
plt.plot(t, V, label='Vaccinated (The Model)', color='red')
plt.plot(t, S_prev, label='Susceptible (Previous Model)', linestyle='--', color='blue')
plt.plot(t, V_prev, label='Vaccinated (Previous Model)', linestyle='--', color='red')
plt.title('Vaccination Impact on Susceptible and Vaccinated Populations')
plt.xlabel('Time (days)')
plt.ylabel('Population')
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# =======================
# Plot 3: Acute and Chronic Infections Comparison
# =======================
plt.figure(figsize=(10, 5))
plt.plot(t, Ac, label='Acute (The Model)', color='cyan')
plt.plot(t, C, label='Chronic (The Model)', color='darkred')
plt.plot(t, Ac_prev, label='Acute (Previous Model)', linestyle='--', color='cyan')
plt.plot(t, C_prev, label='Chronic (Previous Model)', linestyle='--', color='darkred')
plt.title('Acute and Chronic Infection Dynamics')
plt.xlabel('Time (days)')
plt.ylabel('Population')
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
